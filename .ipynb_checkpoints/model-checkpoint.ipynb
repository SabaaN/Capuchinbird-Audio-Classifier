```python
import os
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_io as tfio
```

```python
cf = os.path.join('data', 'Parsed_Capuchinbird_Clips', 'XC3776-3.wav')
Ncf = os.path.join('data', 'Parsed_Not_Capuchinbird_Clips', 'afternoon-birds-song-in-forest-0.wav')
```

```python
#dataloading function

def loadData(fileName):
    #loading encoded wav file
    fileContents=tf.io.read_file(fileName) #this will return a byte encoded string
    #decoding wav 
    wav, sampleRate=tf.audio.decode_wav(fileContents, desired_channels=1)
    #removing the trailing axis (last dimensions of a multi-dimensional array or tensor)
    #it is done simplify the shape of an array or tensor and make it easier to work with
    wav = tf.squeeze(wav,axis=-1) 
    sampleRate = tf.cast(sampleRate,dtype=tf.int64) #Converting the data type of sampleRate to 64-bit integer
    #resampling the amplitude of audio signal to 16000Hz
    wav = tfio.audio.resample(wav,rate_in=sampleRate,rate_out=16000)
    return wav
```

```python
POS=os.path.join('data','Parsed_Capuchinbird_Clips')
NEG=os.path.join('data','Parsed_Not_Capuchinbird_Clips')
```

```python
pos = tf.data.Dataset.list_files(POS+ "\*.wav")
neg = tf.data.Dataset.list_files(NEG+ "\*.wav")

#now add labels and combine positive and negative samples.
#we do this bec we right now dont know what is a positive example and negative example(with no label)

positives = tf.data.Dataset.zip((pos, tf.data.Dataset.from_tensor_slices(tf.ones(len(pos)))))
negatives = tf.data.Dataset.zip((neg, tf.data.Dataset.from_tensor_slices(tf.zeros(len(neg)))))
data=positives.concatenate(negatives)
```

```python
#now we will calculate wave cycle Length
lengths = []
for file in os.listdir(os.path.join('data', 'Parsed_Capuchinbird_Clips')):
    tensor_wave = loadData(os.path.join('data', 'Parsed_Capuchinbird_Clips', file))
    lengths.append(len(tensor_wave))
```

```python
#now building the data preporcessing function
def preprocess(file_path, label): 
    wav = loadData(file_path)
    wav = wav[:48000]
    zero_padding = tf.zeros([48000] - tf.shape(wav), dtype=tf.float32)
    wav = tf.concat([zero_padding, wav],0)
    spectrogram = tf.signal.stft(wav, frame_length=320, frame_step=32)
    spectrogram = tf.abs(spectrogram)
    spectrogram = tf.expand_dims(spectrogram, axis=2)
    return spectrogram, label
```

```python
filepath, label = positives.shuffle(buffer_size=10000).as_numpy_iterator().next()
```

```python
spectrogram, label = preprocess(filepath, label)
```

```python
plt.figure(figsize=(30,20))
plt.imshow(tf.transpose(spectrogram)[0])
plt.show()
```

```python
#now creating training and testing partitions
# creating the pipeline 
data = data.map(preprocess)
data = data.cache()
data = data.shuffle(buffer_size=1000)
data = data.batch(16)
data = data.prefetch(8)
```

```python
train = data.take(36)
test = data.skip(36).take(15)
```

```python
#testing just one batch
samples, labels = train.as_numpy_iterator().next()
```

```python
samples.shape
```

```python
#now building the deep learning model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Flatten
```

```python
model = Sequential()
model.add(Conv2D(16, (3,3), activation='relu', input_shape=(1491, 257,1)))
model.add(Conv2D(16, (3,3), activation='relu'))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('Adam', loss='BinaryCrossentropy', metrics=[tf.keras.metrics.Recall(),tf.keras.metrics.Precision()])
```

```python
model.summary()
```

```python
#for test, running 4 epochs, 
hist = model.fit(train, epochs=4, validation_data=test)
```

```python
#now loading up mp3s
def loadingData(filename):
    """ Load a WAV file, convert it to a float tensor, resample to 16 kHz single-channel audio. """
    res = tfio.audio.AudioIOTensor(filename)
    # Convert to tensor and combine channels 
    tensor = res.to_tensor()
    tensor = tf.math.reduce_sum(tensor, axis=1) / 2 
    # Extract sample rate and cast
    sample_rate = res.rate
    sample_rate = tf.cast(sample_rate, dtype=tf.int64)
    # Resample to 16 kHz
    wav = tfio.audio.resample(tensor, rate_in=sample_rate, rate_out=16000)
    return wav
```

```python
mp3 = os.path.join('data', 'Forest Recordings', 'recording_00.mp3')
```

```python
wav = loadingData(mp3)
audio_slices = tf.keras.preprocessing.timeseries_dataset_from_array(
    wav, wav, sequence_length=48000, sequence_stride=48000, batch_size=1
)
```

```python
samples, index = audio_slices.as_numpy_iterator().next()
```

```python
#this function converts clips into windowed spectrograms
def preprocess_mp3(sample, index):
    sample = sample[0]
    zero_padding = tf.zeros([48000] - tf.shape(sample), dtype=tf.float32)
    wav = tf.concat([zero_padding, sample],0)
    spectrogram = tf.signal.stft(wav, frame_length=320, frame_step=32)
    spectrogram = tf.abs(spectrogram)
    spectrogram = tf.expand_dims(spectrogram, axis=2)
    return spectrogram
```

```python
#Converting longer clips into windows, making predictions
audio_slices = tf.keras.preprocessing.timeseries_dataset_from_array(wav, wav, sequence_length=16000, sequence_stride=16000, batch_size=1)
audio_slices = audio_slices.map(preprocess_mp3)
audio_slices = audio_slices.batch(64)
```

```python
res = model.predict(audio_slices)
res = [1 if prediction > 0.5 else 0 for prediction in res]
```

```python
#Grouping consecutive detections
from itertools import groupby

res = [key for key, group in groupby(res)]
calls = tf.math.reduce_sum(res).numpy()
```

```python
calls
```

```python
#Looping over all recordings and make predictions
results = {}
for file in os.listdir(os.path.join('data', 'Forest Recordings')):
    FILEPATH = os.path.join('data','Forest Recordings', file)
    
    wav = loadingData(FILEPATH)
    audio_slices = tf.keras.preprocessing.timeseries_dataset_from_array(wav, wav, sequence_length=48000, sequence_stride=48000, batch_size=1)
    audio_slices = audio_slices.map(preprocess_mp3)
    audio_slices = audio_slices.batch(64)
    
    res = model.predict(audio_slices)
    
    results[file] = res
```

```python
results
```

```python
#converting predictions into classes
class_preds = {}
for file, logits in results.items():
    class_preds[file] = [1 if prediction > 0.99 else 0 for prediction in logits]
#class_preds
```

```python
#Grouping consecutive detections
postprocessed = {}
for file, scores in class_preds.items():
    postprocessed[file] = tf.math.reduce_sum([key for key, group in groupby(scores)]).numpy()
# postprocessed
```

```python
import csv
with open('results.csv', 'w', newline='') as f:
    writer = csv.writer(f, delimiter=',')
    writer.writerow(['recording', 'capuchin_calls'])
    for key, value in postprocessed.items():
        writer.writerow([key, value])
```
